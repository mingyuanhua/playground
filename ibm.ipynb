```python
# Top K Frequent Words
from typing import List
from collections import Counter
from heapq import heapify, heappop

class Solution:
    def topKFrequent(self, words: List[str], k: int) -> List[str]:
        cnt = Counter(words)
        heap = [(-freq, word) for word, freq in cnt.items()]
        heapify(heap)

        return [heappop(heap)[1] for _ in range(k)]
    
# Input: words = ["the","day","is","sunny","the","the","the","sunny","is","is"], k = 4
# Output: ["the","is","sunny","day"]
# Explanation: "the", "is", "sunny" and "day" are the four most frequent words, with the number of occurrence being 4, 3, 2 and 1 respectively.
```

```python
# Merge Intervals
# Assumption: non-overlapping
class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        # sort the interval by its first element
        intervals.sort(key=lambda x: x[0])

        # merge it if start of cur interval >= cur merged end
        # append to the merged list if <
        merged = []
        for interval in intervals:
            if not merged or merged[-1][1] < interval[0]:
                merged.append(interval)
            else:
                merged[-1][1] = max(merged[-1][1], interval[1])

        return merged
    
# Input: intervals = [[1,3],[2,6],[8,10],[15,18]]
# Output: [[1,6],[8,10],[15,18]]
# Explanation: Since intervals [1,3] and [2,6] overlap, merge them into [1,6].
```

```python
# Reverse Integer
"""
Reverse interger
1. record the sign
2. reverse the positive one
3. add the sign
4. if out of int bounds, then return 0
"""
class Solution:
    def reverse(self, x: int) -> int:
        res = 0
        sign = 1 if x >= 0 else -1
        x = x * sign
        while x > 0:
            res = res * 10 + x % 10
            x //= 10
        res *= sign
        return res if res >= -(2**31) and res <= (2**31 - 1) else 0

# Input: x = -123
# Output: -321
```

```python
# Add Two Numbers
from typing import Optional
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next  

class Solution:
    # head is the last digit of the integer
    # use a dummy head trick, and use cur for iteration
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode(0)
        cur = dummy
        carry = 0
        while l1 != None or l2 != None or carry != 0:
            l1Val = l1.val if l1 else 0
            l2Val = l2.val if l2 else 0
            sum = l1Val + l2Val + carry
            carry = sum // 10
            newNode = ListNode(sum % 10)
            cur.next = newNode
            cur = newNode
            l1 = l1.next if l1 else None
            l2 = l2.next if l2 else None
        return dummy.next


# Input: l1 = [9,9,9,9,9,9,9], l2 = [9,9,9,9]
# Output: [8,9,9,9,0,0,0,1]
```

```python
# Swap Nodes in Pairs

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def swapPairs(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # use recursion to solve this problem
        # base case
        if not head or not head.next:
            return head
        # recursive rule
        first_node = head
        second_node = head.next
        # swap
        first_node.next = self.swapPairs(second_node.next)
        second_node.next = first_node
        return second_node


class Solution:
    def swapPairs(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # Optimize O(n) O(n) to O(n) O(1)
        # use iterative to solve this problem
        # use a dummy head and a prev node and cur node
        dummy = ListNode(-1)
        dummy.next = head
        prev = dummy
        cur = head

        while cur and cur.next:
            # swap
            first_node = cur
            second_node = cur.next
            # p -> 1 -> 2 -> 3 -> 4
            prev.next = second_node
            first_node.next = second_node.next
            second_node.next = first_node
            
            prev = first_node
            cur = first_node.next

        return dummy.next
```

```python
# First Missing Positive
# solution 1
class Solution:
    def firstMissingPositive(self, nums: List[int]) -> int:
        n = len(nums)

        # corner case
        if 1 not in nums:
            return 1
        
        # replace special values
        for i in range(n):
            if nums[i] <= 0 or nums[i] > n:
                nums[i] = 1

        for i in range(n):
            a = abs(nums[i])
            if a == n:
                nums[0] = - abs(nums[0])
            else:
                nums[a] = - abs(nums[a])

        for i in range(1, n):
            if nums[i] > 0:
                return i

        if nums[0] > 0:
            return n

        return n + 1
    
```
